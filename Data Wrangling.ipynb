## b. Tidiness Issues

- Phone + email in contact col `patients`

- Merge the `adverse` table to `treatment` table

#### Define

```python
# Phone and email in separate cols
```

```python
# Novodra and Aurolin using the melt function
```

#### Code

```python
patients_1=patients_copy[patients_copy['address'].isnull()]
patients_copy.dropna(inplace=True)
```

```python
emails=[]
numbers=[]
for i in patients_copy['contact']:
    pattern=re.compile(r'[a-zA-Z][a-zA-Z0-9.-]+[@][a-zA-Z0-9-]+\.[a-zA-Z0-9.-]+[a-zA-Z]')
    numbers.append(i.replace(pattern.findall(i)[0],''))
    emails.append(pattern.findall(i)[0])
```

```python
patients_copy['Emails']=emails
patients_copy['Phone numbers']=numbers

patients_copy=pd.concat([patients_copy,patients_1]).reset_index()
```

```python
patients_copy=patients_copy.drop(columns=['index','contact'])
```

```python
patients_copy
```

```python
treatments_final=pd.melt(treatments_final, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'],
                           var_name='treatment', value_name='dose')
```

```python
treatments_final=treatments_final[treatments_final['dose']!='-']
```

```python
treatments_final['start'],treatments_final['end']=treatments_final['dose'].str.split('-').str
```

```python
treatments_final.drop(columns={'dose'}, inplace=True)
```

```python
treatments_final['start']=treatments_final['start'].str.replace('u','')
treatments_final['end']=treatments_final['end'].str.replace('u','')

treatments_final['start']=treatments_final['start'].astype('int')
treatments_final['end']=treatments_final['end'].astype('int')

treatments_final.reset_index(drop=True,inplace=True)
```

### Assessing Data

In this step, the data is to be understood more deeply. Before implementing methods to clean it, you will definitely need to have a better idea about what the data is about.

### Types of Unclean Data

There are 2 kinds of unclean data

1. <u>Dirty Data</u> (Data with Quality issues):<br>
    Dirty data, also known as low quality data. Low quality data has content issues.<br>
2. <u>Messy Data</u> (Data with tidiness issues)<br>
    Messy data, also known as untidy data. Untidy data has structural issues.

### 1. Dirty Data

When the data has following kinds of issues:

1. Duplicated data
2. Missing Data
3. Corrupt Data
4. Inaccurate Data

This kind of data is known as Dirty data

### 2. Messy Data

When the data has structural issue is it known as Messy or untidy data.

Tidy data has the following properties:

1. Each variable forms a column
2. Each observation forms a row
3. Each observational unit forms a table

If a data violates any one of the above it is known as untidy data
